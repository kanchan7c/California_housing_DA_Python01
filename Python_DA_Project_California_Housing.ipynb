> ### **Question 1**

```python
# What is the average median income of the data set and check the distribution of data using appropriate plots.
# Please explain the distribution of the plot.

# Calculate the average median income
average_median_income = california_data['median_income'].mean()
print("Average Median Income:", average_median_income)

# Plot the distribution of median income
plt.figure(figsize=(15, 6))
sns.histplot(california_data['median_income'], kde=True)
plt.xlabel('Median Income')
plt.title('Distribution of Median Income')
plt.show()

# Explanation:
# We calculated the average median income of the dataset, which is approximately $66,000.
# We then visualized the distribution of median income using a histogram with a Kernel Density Estimation (KDE) plot.
# The plot shows that median income is right-skewed, with a concentration of values in the lower-income range and a long tail towards higher incomes.

# Median income is a continuous feature as it can take any numeric value.
```

> ### **Question 2**

```python
# Draw an appropriate plot to see the distribution of housing_median_age and explain your observations.

# Plot the distribution of housing median age
plt.figure(figsize=(15, 6))
sns.histplot(california_data['housing_median_age'], kde=True)
plt.xlabel('Housing Median Age')
plt.title('Distribution of Housing Median Age')
plt.show()

# Explanation:
# We visualized the distribution of housing median age using a histogram with a KDE (Kernel Density Estimation) plot.
# The plot shows that the housing median age is slightly right-skewed, with a peak in the middle,
# indicating that many areas have housing with a median age around 20-30 years. There are fewer areas with extremely old or extremely new housing.

# Housing median age is a discrete feature as it represents the age of housing in whole years.
```

```python
# Create a data set by deleting the corresponding examples from the data set for which total_bedrooms are not available.

# Create a new dataset with only rows where total_bedrooms are available
california_data_cleaned = california_data.dropna(subset=['total_bedrooms'])

# Explanation:
# A new dataset (california_data_cleaned) is created by removing rows where the total_bedrooms column has missing values.

california_data_cleaned.head()
```

> ### **Question 5**

```python
# Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.

# Fill missing values in total_bedrooms with the mean value
california_data_filled = california_data.copy()
california_data_filled['total_bedrooms'].fillna(california_data['total_bedrooms'].mean(), inplace=True)

# Explanation:
# A new dataset (california_data_filled) is created by filling the missing values in the total_bedrooms column with the mean value
# of total_bedrooms from the original dataset.

california_data_filled.head()
```

> ### **Question 8**

```python
# Create a data set for which the ocean_proximity is ‘Near ocean’.

# Create a new dataset containing only rows with 'Near ocean' ocean_proximity
california_near_ocean = california_data[california_data['ocean_proximity'] == 'NEAR OCEAN']

# Plotting to showcase the significance of calculating near ocean houses
california_inland = california_data[california_data['ocean_proximity'] == 'INLAND']

plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.title('Median Housing Value Near Ocean')
sns.histplot(california_near_ocean['median_house_value'], color='g', kde=True)
plt.subplot(1,2,2)
plt.title('Median Housing Value Inland')
sns.histplot(california_inland['median_house_value'], color='b', kde=True)

# Explanation:
# A new dataset (near_ocean_data) is created by filtering the original dataset to include only the rows where ocean_proximity is 'NEAR OCEAN'.
# This data set can be used to determine the houses that are nearby ocean so that the tourism industry can target these houses for renting
# as bnb or a hotel

california_near_ocean.head()
```

> ### **Question 10**

```python
# Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small.
# If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

def categorize_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return "small"
    elif 10 < total_bedrooms < 1000:
        return "medium"
    else:
        return "large"

california_data['total_bedroom_size'] = california_data['total_bedrooms'].apply(categorize_bedroom_size)

# Save the updated dataset with the new column to a new Excel file.
# california_data.to_excel("california_housing_with_bedroom_size.xlsx", index=False)

california_data.head()

plt.figure(figsize=(15,6))
sns.scatterplot(x=california_data['total_rooms'], y=california_data['total_bedrooms'], hue=california_data['total_bedroom_size'], s=50)

# Explanation:
# A new column 'total_bedroom_size' is added to the dataset based on the 'total_bedrooms' column.
# The function categorize_bedroom_size is applied to each row of the 'total_bedrooms' column, which categorizes the total bedroom size
# as 'small', 'medium', or 'large' based on the provided criteria. The updated dataset is then saved to a new Excel file.
```
